## Stage 2:
### 2) Calculate mutual_info_score of target variable created in stage 1 & ALL non funding variables and share on a jupyter notebook.

To calculate mutual_info_score, we need to cluster the data. I will use K-Means clustering for this. To perform clustering we will need to:
1. Fill in null values for both categorical and numeric data
2. one-hot encode the categorical variables

Let's start by identifying categorical and numeric columns.

```python
# print out all columns included in data ('Brief Description and 'Index Number' were dropped 
# before)
{column[0] for column in data.columns}
```

```python
# print out the columns that have any numerical data and store inside num_columns
non_cat_index = []

for column in data.columns:
    if np.sum(pd.to_numeric(data[column], errors='coerce').notnull()) != 0:
        print(column)
        non_cat_index.append(column)
```

After manually going through the columns above, all were found to contain either numeric values or NaNs and do are numerical columns.

```python
# create a column index containing only categorical columns
cat_index = data.columns.difference(non_cat_index)
# create a column index containing only numerical columns minus the 3 funding columns as
# per problem statement
num_index = pd.MultiIndex.from_tuples(non_cat_index).difference([('Funding FY2008', ''),
                                      ('Funding FY2009', ''),
                                      ('Funding FY2010', ''),
                                     ])
```

```python
# manually checking to see if these are the categorical and numerical columns we want
from IPython.display import display

display("categorical indices", list(cat_index), "numerical indices", list(num_index))
```

```python
# CategoricalEncoder copied from Issue PR #9151 and not my work. The code is very long so
# I put it inside a module for presentation purposes
from data_transformers.transformers import CategoricalEncoder 
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler

stage2_num_pipeline = Pipeline([
    ("data_selector", DataSelector(columns=num_index)),
    ("make_numeric", MakeNumeric()),
    ("fill_na", FillNA(fill_with=0)),
    ("std_scaler", StandardScaler()),
])

stage2_cat_pipeline = Pipeline([
    ("data_selector", DataSelector(columns=cat_index)),
    ("fill_na", FillNA(fill_with='NA')),
    ("categorical_encoder", CategoricalEncoder(encoding='onehot-dense')),
    ("std_scaler", StandardScaler()),
])

stage2_full_pipeline = FeatureUnion([
    ('num_pipeline', stage2_num_pipeline),
    ('cat_pipeline', stage2_cat_pipeline),
])
```

```python
# transform the data for stage 2
data_s2 = stage2_full_pipeline.fit_transform(data.copy())
```

```python
# use K-Means to generate the clustering
from sklearn.cluster import KMeans

kmeans = KMeans()
kmeans.fit_transform(data_s2)
```

```python
from sklearn.metrics import mutual_info_score

mutual_info_score(labels.values , kmeans.labels_)
```

A mutual info score close to zero implies that knowing which cluster an instance belongs to does not provide new information about what its class might be. 

This means that we can expect low values for prediction accuracy or ROC area under the curve.

## Stage 3:
### 1) Divide data into train & test samples. (70-30 split)

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

# store the data processed for stage 2 into features
features = data_s2
# we already placed the target variable, Funding_Increase? into the variable labels which we 
# use below
features_train, features_test, labels_train, labels_test = train_test_split(features, 
                                                                             labels, 
                                                                             test_size=0.3)
```

```python
# are the labels in the training data balanced?
labels_train.value_counts()
```

The two classes (1 and 0) are split 57 % to 43 % which is not very unbalanced. Accuracy can be a good estimate of performance.

## Stage 3:
### 2) Select features & build xgboost model. 

```python
# let's first visualize the data in 2D through dimensionality reduction (using PCA) followed by 
# visualization (using t-SNE)
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# calculate the principal components explaining 90 % of the variance. Higher the percentage
# the more number of components will be generated
pca = PCA(n_components=0.90)
features_train_pca = pca.fit_transform(features_train)

# apply t-SNE with 2 components so that the result can be visualized
tsne = TSNE(n_components=2)
features_train_tsne = tsne.fit_transform(features_train_pca)
```

```python
for i in [0, 1]:
    plt.scatter(features_train_tsne[(labels_train == i), 0], 
                features_train_tsne[(labels_train == i), 1],
                label=i)
plt.legend(loc='lower right')
plt.title("2D visualization of Funding Increase? variable")
plt.xlabel("First Component")
plt.ylabel("Second Component");
```

The 2D t-SNE visualization does not show any distinct clusters of data and this agrees with the low mutual info score of ~ 0.009. It does not appear that a simple decision boundary can separate the data. 

I expect that the accuracy of model prediction would be close to the accuracy of random choice (.50).

```python
# train a XGBClassifier
clf = XGBClassifier(early_stopping_rounds=5)
clf.fit(features_train, labels_train)
labels_pred = clf.predict(features_train)
```

```python
# calculate the accuracy score when predicting training data
from sklearn.metrics import accuracy_score, roc_auc_score

accuracy_score(labels_train, labels_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import time

t0 = time.time()
param_dist = {
    'max_depth': [1, 3, 9],
    'n_estimators': [3000],
    'learning_rate': [0.03, 0.1, 0.3, 1],
}

random_search = RandomizedSearchCV(clf, 
                                   param_distributions=param_dist,
                                   n_iter=10)

random_search.fit(features_train, labels_train)
t1 = time.time()
print("Time taken: {}s".format(t1-t0))
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

```python
best_xgb_clf = random_search.best_estimator_
```

```python
labels_pred = best_xgb_clf.predict(features_test)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(labels_test, labels_pred)
```
